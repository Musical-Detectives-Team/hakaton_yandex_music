## Введение

**Это вторая из трех тетрадей: EDA, PREprocessing, MODeling.**

**Задача:**

В этом хакатоне вам предлагается разработать решение, которое:

- **Первая задача:** может классифицировать треки по признаку кавер-некавер;
- **Вторая задача:** связывать (группировать) каверы и исходный трек;
- **Третья задача:** находит исходный трек в цепочке каверов.

**Данные:**

Файл covers.json содержит **разметку каверов**, сделанную редакторами сервиса:

- track_id - уникальный идентификатор трека;
- track_remake_type - метка, присвоенная редакторами. Может принимать значения ORIGINAL и COVER;
- original_track_id - уникальный идентификатор исходного трека.

💡 Обратите внимание, что не для всех каверов известны идентификаторы исходных треков!!!

**Метаинформация**

- track_id - уникальный идентификатор трека;
- dttm - первая дата появления информации о треке;
- title - название трека;
- language - язык исполнения;
- isrc - международный уникальный идентификатор трека;
- genres - жанры;
- duration - длительность трека;

**Текст песен**

- track_id - уникальный идентификатор трека;
- lyricId - уникальный идентификатор текста;
- text - текст трека.

### `genres`

#### Укрупнение категорий

Посмотрим какие категории у нас есть.

```python
df_united_task_1['genres'].value_counts().head(50)
```

У нас слишком много категорий жанров. При таком количестве информация становится почти бесполезной.

Вручную выделим крупные категории. Запишем их в порядке убывания количества треков.

```python
# крупные категории
genres_big_categories = ['FOLK', 'POP', 'ROCK', 'ALTERNATIVE', 'ELECTRINICS', 'SOUNDTRACK',
                        'SPIRITUAL', 'RAP', 'CLASSICAL', 'JAZZ', 'METAL', 'BARD', 'COUNTRY',
                        'PUNK', 'RELAX', 'INDIE', 'CHILDREN', 'RAGGAE',
                        'RNB', 'BLUES']
```

У нас получилось крупных категорий:

```python
len(genres_big_categories)
```

21 категорией будет 'OTHER'.

В эту же категорию будут записаны треки без категорий(пока не будет придумано, как восстановить пропущенные жанры)

Создадим признак `genres_big_cat`, в котором каждому треку будет присвоен самый популярный трек, к которому его отнесли. И этот жанр будет только один.

```python
df_united_task_1.reset_index(inplace=True, drop=True)

# построчно разберемся с каждым жанром
for i in range(len(df_united_task_1['genres'])):
#     переведем из списков в строки
    string = ''
    for k in df_united_task_1.loc[i, 'genres']:
        string += k + ' '
    # уберем последний пробел и запишем обратно в ячейку
        df_united_task_1.loc[i, 'genres'] = string[:-1]
   
    stop = False
#     проходим по всем 20 категориям, если есть совпадение, то записываем первую самую важную
    for j in range(len(genres_big_categories)):
        if genres_big_categories[j] in string:
            if ~stop:
#         записываем genres_big_categories[j] как категорию
                df_united_task_1.loc[i, 'genres_big_cat'] = genres_big_categories[j]
#                 display(genres_big_categories[j])
                stop = True     
#     если не нашли одну из 20 категорий, то записываем OTHER
    if (j == len(genres_big_categories)-1) & ~stop: 
#         display('OTHER')
        df_united_task_1.loc[i, 'genres_big_cat'] = 'OTHER'
```
