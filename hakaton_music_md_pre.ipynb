–£ –Ω–∞—Å —Å–ª–∏—à–∫–æ–º –º–Ω–æ–≥–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–π –∂–∞–Ω—Ä–æ–≤. –ü—Ä–∏ —Ç–∞–∫–æ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –ø–æ—á—Ç–∏ –±–µ—Å–ø–æ–ª–µ–∑–Ω–æ–π.

–í—Ä—É—á–Ω—É—é –≤—ã–¥–µ–ª–∏–º –∫—Ä—É–ø–Ω—ã–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏. –ó–∞–ø–∏—à–µ–º –∏—Ö –≤ –ø–æ—Ä—è–¥–∫–µ —É–±—ã–≤–∞–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Ç—Ä–µ–∫–æ–≤.

```python
# –∫—Ä—É–ø–Ω—ã–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
genres_big_categories = ['FOLK', 'POP', 'ROCK', 'ALTERNATIVE', 'ELECTRINICS', 'SOUNDTRACK',
                        'SPIRITUAL', 'RAP', 'CLASSICAL', 'JAZZ', 'METAL', 'BARD', 'COUNTRY',
                        'PUNK', 'RELAX', 'INDIE', 'CHILDREN', 'RAGGAE',
                        'RNB', 'BLUES']
```

–£ –Ω–∞—Å –ø–æ–ª—É—á–∏–ª–æ—Å—å –∫—Ä—É–ø–Ω—ã—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π:

```python
len(genres_big_categories)
```

21 –∫–∞—Ç–µ–≥–æ—Ä–∏–µ–π –±—É–¥–µ—Ç 'OTHER'.

–í —ç—Ç—É –∂–µ –∫–∞—Ç–µ–≥–æ—Ä–∏—é –±—É–¥—É—Ç –∑–∞–ø–∏—Å–∞–Ω—ã —Ç—Ä–µ–∫–∏ –±–µ–∑ –∫–∞—Ç–µ–≥–æ—Ä–∏–π(–ø–æ–∫–∞ –Ω–µ –±—É–¥–µ—Ç –ø—Ä–∏–¥—É–º–∞–Ω–æ, –∫–∞–∫ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∂–∞–Ω—Ä—ã)

–°–æ–∑–¥–∞–¥–∏–º –ø—Ä–∏–∑–Ω–∞–∫ `genres_big_cat`, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–∞–∂–¥–æ–º—É —Ç—Ä–µ–∫—É –±—É–¥–µ—Ç –ø—Ä–∏—Å–≤–æ–µ–Ω —Å–∞–º—ã–π –ø–æ–ø—É–ª—è—Ä–Ω—ã–π —Ç—Ä–µ–∫, –∫ –∫–æ—Ç–æ—Ä–æ–º—É –µ–≥–æ –æ—Ç–Ω–µ—Å–ª–∏. –ò —ç—Ç–æ—Ç –∂–∞–Ω—Ä –±—É–¥–µ—Ç —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω.

```python
df_united_task_1.reset_index(inplace=True, drop=True)

# –ø–æ—Å—Ç—Ä–æ—á–Ω–æ —Ä–∞–∑–±–µ—Ä–µ–º—Å—è —Å –∫–∞–∂–¥—ã–º –∂–∞–Ω—Ä–æ–º
for i in range(len(df_united_task_1['genres'])):
#     –ø–µ—Ä–µ–≤–µ–¥–µ–º –∏–∑ —Å–ø–∏—Å–∫–æ–≤ –≤ —Å—Ç—Ä–æ–∫–∏
    string = ''
    for k in df_united_task_1.loc[i, 'genres']:
        string += k + ' '
    # —É–±–µ—Ä–µ–º –ø–æ—Å–ª–µ–¥–Ω–∏–π –ø—Ä–æ–±–µ–ª –∏ –∑–∞–ø–∏—à–µ–º –æ–±—Ä–∞—Ç–Ω–æ –≤ —è—á–µ–π–∫—É
        df_united_task_1.loc[i, 'genres'] = string[:-1]
   
    stop = False
#     –ø—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤—Å–µ–º 20 –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º, –µ—Å–ª–∏ –µ—Å—Ç—å —Å–æ–≤–ø–∞–¥–µ–Ω–∏–µ, —Ç–æ –∑–∞–ø–∏—Å—ã–≤–∞–µ–º –ø–µ—Ä–≤—É—é —Å–∞–º—É—é –≤–∞–∂–Ω—É—é
    for j in range(len(genres_big_categories)):
        if genres_big_categories[j] in string:
            if ~stop:
#         –∑–∞–ø–∏—Å—ã–≤–∞–µ–º genres_big_categories[j] –∫–∞–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏—é
                df_united_task_1.loc[i, 'genres_big_cat'] = genres_big_categories[j]
#                 display(genres_big_categories[j])
                stop = True     
#     –µ—Å–ª–∏ –Ω–µ –Ω–∞—à–ª–∏ –æ–¥–Ω—É –∏–∑ 20 –∫–∞—Ç–µ–≥–æ—Ä–∏–π, —Ç–æ –∑–∞–ø–∏—Å—ã–≤–∞–µ–º OTHER
    if (j == len(genres_big_categories)-1) & ~stop: 
#         display('OTHER')
        df_united_task_1.loc[i, 'genres_big_cat'] = 'OTHER'
```
